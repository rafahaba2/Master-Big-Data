### A function to do all this steps

```python
def getTextWaPo(url):
    response = requests.get(url)
    page = response.text
    soup = BeautifulSoup(page, 'html.parser')
    text = ' '.join(map(lambda p: p.text, soup.find_all('article')))
    return text.encode('ascii', errors = 'replace').decode().replace('?', ' ')
```

```python
text = getTextWaPo(articleURL)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

### Tokenizying the text into sentences

```python
sents = sent_tokenize(text)
sents
```

### Tokenizying words and setting them into lowercase

```python
word_sent = word_tokenize(text.lower())
word_sent
```

### Removing stopWords from the text and punctuation

```python
_stopwords = set(stopwords.words('english') + list(punctuation))
```

```python
word_sent = [word for word in word_sent if word not in _stopwords]
```

```python
word_sent
```

### Frequency that words appear in the text

```python
from nltk.probability import FreqDist
```

```python
from nltk.probability import FreqDist
freq = FreqDist(word_sent)
freq
```

```python
from heapq import nlargest
```

### Printing the 10 words that appear more times.

```python
nlargest(10, freq, key = freq.get)
```

```python
from collections import defaultdict 
```

### Sentences ranking 
Ranking of sentences by the frequency of the words contained in them.

```python
from collections import defaultdict 
ranking = defaultdict(int)

for i, sent in enumerate(sents):
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]
```

```python
ranking
```

### List the index of the 4 sentences with higher ranking

```python
sents_idx = nlargest(4, ranking, key = ranking.get)
sents_idx
```

### Print the 4 sentences with higher ranking by index order

```python
[sents[j] for j in sorted(sents_idx)]
```

### Function for summarizing a text

```python
def summarize(text, n):
    #Tokenize sentences
    sents = sent_tokenize(text)
    
    #Tokenize words from sentences
    word_sent = word_tokenize(text.lower())
    
    #Remove stopwords
    _stopwords = set(stopwords.words('english') + list(punctuation))
    word_sent = [word for word in word_sent if word not in _stopwords]
    
    #stemming
    from nltk.stem.lancaster import LancasterStemmer
    st = LancasterStemmer()
    stemmedWords = [st.stem(word) for word in word_tokenize(text)]
    
    #Frequency words appear in sentences
    freq = FreqDist(stemmedWords)
    
    
    ranking = defaultdict(int)
    
    for i, sent in enumerate(sents):
        for w in word_tokenize(sent.lower()):
            #stemming
            if w in freq:
                ranking[i] += freq[w]
    
    sents_idx = nlargest(4, ranking, key = ranking.get)        
    return [sents[j] for j in sorted(sents_idx)]
```
